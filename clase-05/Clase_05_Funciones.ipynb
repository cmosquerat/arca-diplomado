## 1. El Dolor: Código Repetido

Tres clientes, el mismo cálculo copiado tres veces:

````python
cajas_1, precio_1 = 30, 12.50
ingreso_1 = cajas_1 * precio_1
print(f"Cliente 1: {ingreso_1}")

cajas_2, precio_2 = 15, 12.50
ingreso_2 = cajas_2 * precio_2
print(f"Cliente 2: {ingreso_2}")

cajas_3, precio_3 = 50, 12.50
ingreso_3 = cajas_3 * precio_3
print(f"Cliente 3: {ingreso_3}")
````

Si el jefe dice *"ahora el precio incluye IVA"*, hay que cambiar **3 lugares**.  
Con 50 clientes serían 50 bloques idénticos. Las **funciones** resuelven esto.

## 2. Anatomía de una Función

```
def calcular_ingreso(cajas, precio):   ← firma: def + nombre + parámetros
    ingreso = cajas * precio            ← cuerpo (indentado)
    return ingreso                      ← salida
```

- `def` inicia la definición
- El nombre describe **qué hace** (verbo + sustantivo)
- Los **parámetros** son las entradas
- `return` devuelve el resultado. Sin él, la función devuelve `None`

````python
def calcular_ingreso(cajas, precio):
    ingreso = cajas * precio
    return ingreso

resultado = calcular_ingreso(30, 12.50)
print("Ingreso:", resultado)
````

Ahora el mismo cálculo de arriba se reduce a:

````python
print(f"Cliente 1: {calcular_ingreso(30, 12.50)}")
print(f"Cliente 2: {calcular_ingreso(15, 12.50)}")
print(f"Cliente 3: {calcular_ingreso(50, 12.50)}")
````

## 3. Los Parámetros Son la Pieza Clave

Sin parámetros, una función siempre haría lo mismo. **Con ellos, es reutilizable.**

````python
# Sin parámetros: siempre hace lo mismo
def ingreso_fijo():
    return 30 * 12.50

print(ingreso_fijo())
print(ingreso_fijo())

# Con parámetros: flexible
def ingreso(cajas, precio):
    return cajas * precio

print(ingreso(30, 12.50))
print(ingreso(15, 9.00))
````

### El orden importa — Parámetros posicionales

Python asigna los valores en el **orden** en que los pasan. Si intercambian el orden, puede dar error o resultados incorrectos.

Pueden evitar esto pasando argumentos **por nombre** (`precio=12.50`):

````python
def describir(producto, cantidad, precio):
    total = cantidad * precio
    return f"{producto}: {cantidad} x {precio} = {total}"

# Correcto: en orden
print(describir("Coca-Cola", 30, 12.50))

# Por nombre: el orden no importa
print(describir(precio=12.50, cantidad=30, producto="Coca-Cola"))
````

### ¿Qué pasa si paso más o menos argumentos?

Python es estricto: la cantidad de argumentos debe coincidir con los parámetros.

## 5. `return` vs `print` — No Son lo Mismo

````python
def mostrar_total(cajas, precio):
    print(cajas * precio)

resultado = mostrar_total(30, 12.50)
print("Guardado:", resultado)
````

````python
def calcular_total_v2(cajas, precio):
    return cajas * precio

resultado = calcular_total_v2(30, 12.50)
print("Guardado:", resultado)
print("Con IVA:", resultado * 1.12)
````

`print()` muestra texto en pantalla. `return` entrega un valor que el programa puede guardar, operar o pasar a otra función.

## 10. Funciones que Procesan Colecciones

````python
def resumen_ventas(ventas, meta):
    total = 0
    sobre_meta = 0
    for v in ventas:
        total += v
        if v >= meta:
            sobre_meta += 1
    return total, sobre_meta

datos = [150, 80, 200, 45, 310, 60]
t, s = resumen_ventas(datos, 100)
print(f"Total: {t}")
print(f"Sobre meta: {s}")
````

## 13. Validación de Parámetros — Funciones que se Protegen

¿Qué pasa cuando los datos son incorrectos? Sin validación, la función puede dar resultados absurdos.

````python
# Sin protección: resultados absurdos
def ingreso_sin_validar(cajas, precio):
    return cajas * precio

print(ingreso_sin_validar(-5, 12.50))
print(ingreso_sin_validar(30, 0))
print(ingreso_sin_validar(30, -10))
````

### Patrón Guardia — Validar antes de calcular

Las validaciones van **al inicio**: si algo está mal, la función se detiene temprano con un `return`.

````python
def procesar_entrega(cliente, pedido, carga, precio):
    if pedido <= 0:
        return {"error": "pedido invalido"}
    if pedido > carga:
        return {"error": "sin stock"}
    if precio <= 0:
        return {"error": "precio invalido"}

    ingreso = pedido * precio
    return {
        "cliente": cliente,
        "ingreso": ingreso,
        "restante": carga - pedido
    }

print(procesar_entrega("Sol", 30, 100, 12.5))
print(procesar_entrega("Sol", -5, 100, 12.5))
print(procesar_entrega("Sol", 30, 10, 12.5))
````

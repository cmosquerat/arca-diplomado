## 0. Antes de Empezar: ¿Qué Es un Dato?

### Datos ≠ Información

Los datos se pueden encontrar "fácilmente" en todos lados:
- Evolución del precio de acciones de una empresa en bolsa
- Estadísticas de resultados deportivos
- Históricos de consumo de ciertos productos
- Precios de mercado de bienes y servicios

La **información**, sin embargo, hay que saber cómo y dónde buscarla:
- Normalmente subyace escondida detrás de los datos
- Obtenerla requiere del **procesamiento y del análisis**
- *Soft information* vs *Hard information*

### Pirámide DIKW: Data → Information → Knowledge → Wisdom

| Nivel | Definición | Ejemplo |
|-------|-----------|--------|
| **Data** | Tener las cifras en crudo de un fenómeno | Registros de compras: artículos, hora, género, edad |
| **Information** | Extraer relaciones, dependencias, causas y consecuencias | Alta correlación: compradores hombres + 5pm-7pm + pañales + cervezas |
| **Knowledge** | Saber cómo hacer frente a la información obtenida | Los padres, después del trabajo, compran pañales y cervezas |
| **Wisdom** | Tener el poder para actuar sobre ese conocimiento | Implementar nuevas estrategias: ubicar cervezas cerca de pañales |

> A medida que subimos la pirámide: **más valor, menos volumen**.

### 🍺🧷 El caso de la cerveza y los pañales (Walmart)

En una cadena de almacenes (Walmart) analizaron los datos de compras de sus clientes:

- **Data:** Los registros de los artículos comprados, junto con datos de la hora, el género del comprador y la edad.
- **Information:** Se descubrió una alta correlación entre: *compradores hombres*, *compras entre 5pm y 7pm*, *pañales* y *cervezas*.
- **Knowledge:** Los padres, después de salir del trabajo, suelen comprar pañales y también cervezas.
- **Wisdom:** Implementar nuevas estrategias de publicidad y mercadeo — ubicar cervezas cerca de pañales.

**Este es el poder de convertir datos en decisiones.** Exactamente lo que aprenderemos en este diplomado.

## 2. DataFrame vs Series

pandas tiene dos estructuras fundamentales:

| Estructura | Dimensiones | Analogía |
|-----------|------------|----------|
| **DataFrame** | 2D (filas × columnas) | Tabla completa / Hoja de Excel |
| **Series** | 1D (una sola columna) | Una columna de Excel |

Un DataFrame es, en esencia, **un diccionario de Series** que comparten el mismo índice.

### Regla práctica

| Sintaxis | Resultado |
|---------|----------|
| `df["col"]` | **Series** (una columna) |
| `df[["col1", "col2"]]` | **DataFrame** (subconjunto) |

> La diferencia importa: algunas funciones solo aceptan DataFrame, otras solo Series.

````python
# Series: una sola columna
s = df["Total"]
print("Tipo:", type(s).__name__)
print("Forma:", s.shape)
print()
s.head()
````

````python
# DataFrame: varias columnas (doble corchete)
sub = df[["Branch", "Total"]]
print("Tipo:", type(sub).__name__)
print("Forma:", sub.shape)
print()
sub.head()
````

## 3. Tipos de Dato en pandas

Cada columna tiene un **dtype** (tipo de dato). Si el tipo está mal, las operaciones fallan **silenciosamente**.

| dtype | Qué guarda | Ejemplo | Operaciones típicas |
|-------|-----------|---------|--------------------|
| `int64` | Enteros | 30, 150 | Suma, media |
| `float64` | Decimales | 12.50 | Suma, media |
| `object` | Texto | "Ana", "Norte" | Conteo, filtro |
| `bool` | Verdadero/Falso | True, False | Filtro |
| `datetime64` | Fechas | 2026-03-25 | Rango, mes, día |
| `category` | Categorías | "A", "B", "C" | Ahorra memoria |

### ⚠️ ¿Por qué importa Numérico vs Categórico?

| Tipo | Operaciones válidas | Si aplicas al tipo incorrecto... |
|------|-------------------|--------------------------------|
| **Numérico** (int64, float64) | Media, mediana, suma, std, histograma, correlación | `df["sucursal"].mean()` → promedio sin sentido |
| **Categórico** (object, category) | Conteo, frecuencia, moda, gráfico de barras, groupby | `df["precio"].sum()` con object → ¡concatena textos! |

> **Regla:** Si el tipo está mal, el análisis está mal. Siempre verificar con `dtypes` antes de operar.

````python
print("=== Tipos de dato del dataset ===")
print(df.dtypes)
````

````python
df.info()
````

### Observación clave

La columna `Date` aparece como `object` (texto), pero debería ser `datetime64`.  
Esto significa que **no podemos** extraer mes, día o año directamente. Hay que convertirla.

### 3.1 Convertir tipos: Fechas y Categorías

````python
# Ejemplo: las fechas suelen llegar como texto (object)
fechas = pd.Series(["2025-01-15", "2025-02-20", "2025-03-10"])
print("Tipo original:", fechas.dtype)

# Convertir a datetime
fechas_dt = pd.to_datetime(fechas)
print("Tipo nuevo:   ", fechas_dt.dtype)

# Ahora podemos extraer componentes
print("Mes:", fechas_dt.dt.month.tolist())
print("Día:", fechas_dt.dt.day_name().tolist())
````

### 3.2 Convertir tipos: `pd.to_numeric()` y `.astype()`

A veces una columna numérica llega como texto (`object`).  
`pd.to_numeric()` con `errors="coerce"` convierte lo que puede y marca lo inválido como `NaN`.

````python
import numpy as np

# Ejemplo: una Series con datos mixtos
s = pd.Series(["10", "20", "treinta", "40"])
print("Tipo original:", s.dtype)

# to_numeric con errors="coerce": invalidos -> NaN
s_num = pd.to_numeric(s, errors="coerce")
print("\nConvertida:")
print(s_num)
print("\nTipo:", s_num.dtype)
````

### Resumen: Herramientas de conversión

| Función | Cuándo usarla | Manejo de errores |
|---------|--------------|-------------------|
| `pd.to_datetime()` | Texto → fecha | `errors="coerce"` |
| `pd.to_numeric()` | Texto → número | `errors="coerce"` |
| `.astype(int)` | Cambio explícito | Falla si hay inválidos |
| `.astype("category")` | Reducir memoria | Siempre funciona |
| `.astype(str)` | Número → texto | Siempre funciona |

> `errors="coerce"` es tu aliado: convierte lo que puede y marca lo demás como `NaN`.

### 3.3 Separar columnas por tipo: `select_dtypes()`

## 4. Datos Faltantes: NaN

### ¿Qué es NaN?

`NaN` = **Not a Number** = "no hay dato".

Puede ocurrir porque:
- Alguien no llenó el campo
- Un sensor falló
- El dato se perdió en una migración

**Importante:**
- `NaN ≠ 0` (cero es un valor)
- `NaN ≠ ""` (string vacío es un valor)
- `NaN ≠ False` (False es un valor)

> Si calculas la media con NaN, pandas lo **ignora silenciosamente**.  
> Si filtras, las filas con NaN pueden **desaparecer** sin aviso.  
> Siempre hay que **detectar y decidir** qué hacer con ellos.

### 4.1 Detectar NaN: `isna()` y porcentaje

````python
# Conteo de NaN por columna
print("=== NaN por columna ===")
print(df.isna().sum())
````

````python
# Porcentaje de NaN
pct = (df.isna().sum() / len(df) * 100).round(1)
print("=== % faltante (solo columnas con NaN) ===")
print(pct[pct > 0])
````

### 4.2 ¿Cuándo usar Media, Mediana o Moda?

| Estrategia | Cuándo usarla | Ejemplo |
|-----------|--------------|--------|
| **Media** (promedio) | Distribución simétrica, sin outliers extremos | Rating, temperatura |
| **Mediana** (valor central) | Hay outliers o distribución sesgada | Salarios, montos de venta (Total) |
| **Moda** (más frecuente) | Datos categóricos (texto) o discretos | Método de pago, ciudad, sucursal |

> **Regla de oro:** Si hay outliers → mediana. Si es texto → moda. Si es simétrica → media.

### Otras estrategias

| Estrategia | Cuándo usarla |
|-----------|---------------|
| `dropna()` — Eliminar filas | Pocos NaN (< 5%). Perder filas no importa. |
| `fillna(0)` — Casi nunca | Solo si 0 tiene significado real en el contexto. |

````python
# Con outliers: rellenar con mediana
mediana_total = df["Total"].median()
print(f"Mediana de Total: {mediana_total}")
df["Total"] = df["Total"].fillna(mediana_total)

# Sin outliers: rellenar con media
media_rating = df["Rating"].mean()
print(f"Media de Rating: {media_rating:.2f}")
df["Rating"] = df["Rating"].fillna(media_rating)

# Categórica: rellenar con moda (valor más frecuente)
moda_payment = df["Payment"].mode()[0]
print(f"Moda de Payment: {moda_payment}")
df["Payment"] = df["Payment"].fillna(moda_payment)

# Eliminar filas donde una columna crítica sea NaN
# df = df.dropna(subset=["Total"])

print(f"\nNaN restantes: {df.isna().sum().sum()}")
````

## 5. Visualizar para Entender: Histograma y Boxplot

Usaremos **seaborn**, una librería que se construye sobre matplotlib y produce gráficas profesionales con muy poco código.

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
print("seaborn listo ✓")
````

### 5.1 Histograma: ¿Cómo se distribuyen los valores?

Un histograma agrupa los valores en "bins" (contenedores) y cuenta cuántos caen en cada uno.  
Responde: ¿Se acumulan a la izquierda? ¿Hay picos? ¿Hay colas largas?

````python
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(df["Total"], bins=25, color="#C82B40", ax=ax)
ax.set_title("Distribución del Total de Venta", fontsize=13, fontweight="bold")
ax.set_xlabel("Total ($)", fontsize=11)
ax.set_ylabel("Frecuencia", fontsize=11)
plt.tight_layout()
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(df["Rating"], bins=20, color="#2563EB", ax=ax)
ax.set_title("Distribución del Rating", fontsize=13, fontweight="bold")
ax.set_xlabel("Rating", fontsize=11)
ax.set_ylabel("Frecuencia", fontsize=11)
plt.tight_layout()
plt.show()
````

### 5.2 Boxplot: Dispersión y Outliers

**Anatomía del boxplot:**

- La **caja** = 50% central de los datos (Q1 a Q3)
- La **línea dentro** = mediana (Q2)
- Los **bigotes** = datos dentro de 1.5 × IQR
- Los **puntos fuera** = outliers (valores atípicos)

> **IQR** (Rango Intercuartílico) = Q3 − Q1

````python
fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=df, x="Product line", y="Total",
            palette="Set2", ax=ax)
ax.set_title("Total por Línea de Producto", fontsize=13, fontweight="bold")
ax.set_xlabel("Línea de Producto", fontsize=11)
ax.set_ylabel("Total ($)", fontsize=11)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=df, x="Branch", y="Rating",
            palette=["#C82B40", "#2563EB", "#16A34A"], ax=ax)
ax.set_title("Rating por Sucursal", fontsize=13, fontweight="bold")
ax.set_xlabel("Sucursal", fontsize=11)
ax.set_ylabel("Rating", fontsize=11)
plt.tight_layout()
plt.show()
````

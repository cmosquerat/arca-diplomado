## 6. Visualización con seaborn y matplotlib

### Referencia rápida

| Gráfica | Para qué sirve | Código seaborn |
|---------|---------------|----------------|
| **Histograma** | Distribución de UNA variable numérica | `sns.histplot(df["col"], bins=20)` |
| **Boxplot** | Dispersión + outliers, comparar grupos | `sns.boxplot(data=df, x="cat", y="num")` |
| **Barras (conteo)** | Frecuencia de categorías | `sns.countplot(data=df, x="cat")` |
| **Barras (valor)** | Comparar un valor entre categorías | `sns.barplot(data=df, x="cat", y="num")` |

### Tips útiles

```python
fig, ax = plt.subplots(figsize=(10, 5))   # tamaño personalizado
ax.set_title("Título", fontweight="bold")  # título
plt.xticks(rotation=45)                    # rotar etiquetas
plt.tight_layout()                         # ajustar márgenes

# Subdividir barras por otra variable:
sns.countplot(data=df, x="col", hue="otra_col")

# Múltiples gráficas en una figura:
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
sns.histplot(..., ax=axes[0])
sns.boxplot(..., ax=axes[1])
```

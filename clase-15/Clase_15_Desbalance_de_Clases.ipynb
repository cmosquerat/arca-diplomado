## 5. ¿Por qué falla? — El desbalance de clases

### La analogía de la certificación

Imaginen que están preparando un examen de certificación. Les dan 100 preguntas de práctica. Notan que en **76** la correcta es **A**. Solo en 24 es B, C o D. El examen es mañana. ¿Estrategia? **Marcar A en todo**. Resultado: **76/100. Aprobado.** No aprendieron nada — solo explotaron el desbalance.

El modelo hace lo mismo: aprende que "se queda" es la respuesta segura.

### La matriz de confusión como heatmap

`sns.heatmap()` convierte una tabla en un mapa de calor:
- `annot=True`: muestra los números
- `fmt="d"`: formato entero
- `cmap="Reds"`: paleta (para heatmaps se usa `cmap`, no `palette`)

````python
cm = confusion_matrix(y, y_pred)

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds")
plt.ylabel("Real")
plt.xlabel("Prediccion del modelo")
plt.title("Matriz de Confusion - Modelo Base")
plt.show()

print(f"\nVP (detectados): {cm[1,1]}")
print(f"FN (se escapan): {cm[1,0]}")
print(f"De {cm[1,0]+cm[1,1]} que renuncian, solo detectamos {cm[1,1]}.")
````

## 6. Solución 1: `class_weight='balanced'`

### ¿Qué hace internamente?

Normalmente el modelo trata cada error igual. Con `class_weight='balanced'`, sklearn calcula un **peso** para cada clase:

$$w_i = \frac{n}{2 \times n_i}$$

En nuestro caso:
- Peso de "se queda": $\frac{1000}{2 \times 763} = 0.66$ → pesa **menos**
- Peso de "renuncia": $\frac{1000}{2 \times 237} = 2.11$ → pesa **3x más**

El modelo **penaliza más** cuando se equivoca con alguien que renuncia.

````python
modelo_bal = LogisticRegression(max_iter=1000, class_weight="balanced")
modelo_bal.fit(X_s, y)
y_pred_bal = modelo_bal.predict(X_s)

print("=== Modelo Base ===")
print(f"  Precision: {precision_score(y, y_pred):.3f}")
print(f"  Recall:    {recall_score(y, y_pred):.3f}")
print()
print("=== Modelo Balanced ===")
print(f"  Precision: {precision_score(y, y_pred_bal):.3f}")
print(f"  Recall:    {recall_score(y, y_pred_bal):.3f}")
````

## 7. Solución 2: Oversampling — duplicar la minoría

En vez de cambiar el modelo, **cambiamos los datos**: repetimos las filas de la clase minoritaria hasta igualar.

````python
from sklearn.utils import resample

df_may = df_enc[df_enc["Attrition"] == 0]
df_min = df_enc[df_enc["Attrition"] == 1]

df_min_up = resample(df_min, replace=True,
                     n_samples=len(df_may), random_state=42)

df_over = pd.concat([df_may, df_min_up])

print(f"Antes:   {len(df_enc)} filas")
print(f"Despues: {len(df_over)} filas")
print(f"  Se queda: {(df_over['Attrition']==0).sum()}")
print(f"  Renuncia: {(df_over['Attrition']==1).sum()}")
````

## 10. Train/Test Split — Evaluar de verdad

### ¿Por qué no podemos evaluar con los mismos datos?

Imaginen que un profesor les da el examen final para que "practiquen"... y luego les toma **exactamente el mismo examen**. Todos sacan 10. Pero ¿realmente aprendieron? Quizás solo **memorizaron las respuestas**.

Eso es lo que hicimos hoy: entrenar y evaluar con los mismos datos. El modelo puede estar **memorizando** en vez de aprendiendo.

### La analogía del piloto

¿Confiarían en un piloto que solo fue evaluado en las **mismas** rutas donde practicó? Un buen piloto puede volar rutas **que nunca practicó** porque aprendió los **principios**, no memorizó las rutas. Un buen modelo predice bien datos **que nunca vio** porque aprendió los **patrones**, no memorizó las filas.

### La solución: separar los datos

Dividimos los datos en dos partes **antes** de entrenar:
- **Train** (~80%): los ejercicios de práctica — el modelo aprende con estos
- **Test** (~20%): el examen final — evaluamos con datos que el modelo **nunca vio**

El modelo NUNCA ve los datos de test durante el entrenamiento.

````python
from sklearn.model_selection import train_test_split

# Usamos iris para el ejemplo
iris = sns.load_dataset("iris")

# Convertir a clasificacion binaria: "setosa" vs "no setosa"
iris["es_setosa"] = (iris["species"] == "setosa").astype(int)

X_iris = iris[["sepal_length", "sepal_width",
               "petal_length", "petal_width"]]
y_iris = iris["es_setosa"]

print(f"Total: {len(y_iris)} filas")
print(f"Setosa: {y_iris.sum()} ({y_iris.mean()*100:.0f}%)")
````

````python
# Separar en train y test
X_train, X_test, y_train, y_test = train_test_split(
    X_iris, y_iris,
    test_size=0.2,       # 20% para test
    random_state=42)     # para reproducibilidad

print(f"Train: {len(X_train)} filas")
print(f"Test:  {len(X_test)} filas")
````

````python
# Entrenar con train, evaluar con test
modelo_iris = LogisticRegression(max_iter=1000)
modelo_iris.fit(X_train, y_train)       # aprende con train
y_pred_iris = modelo_iris.predict(X_test)  # predice sobre test

print(f"Accuracy en test: {accuracy_score(y_test, y_pred_iris):.3f}")
print(f"Recall en test:   {recall_score(y_test, y_pred_iris):.3f}")
print()
print("Esta evaluacion es JUSTA: el modelo nunca vio los datos de test.")
````

## 2. Práctica guiada: Ventas de Coca-Cola

> Una tienda vende **1 200 unidades** de Coca-Cola 600 ml a **\$1.50** cada una.

Usando solo `print()` y operaciones, calculemos:

1. El **ingreso total** de la venta.
2. Si el costo por unidad es \$0.85, ¿cuál es la **ganancia total**?
3. ¿Cuál es el **margen de ganancia** en porcentaje?

**Pista:** Margen = (precio − costo) / precio × 100

```python
# Solución — Práctica guiada
print(1200 * 1.50)                          # Ingreso total
print(1200 * 1.50 - 1200 * 0.85)            # Ganancia total
print((1.50 - 0.85) / 1.50 * 100)           # Margen %
```

## 3. Práctica autónoma: Recepción de cajas

> Una bodega recibe **500 cajas** de Coca-Cola. Cada caja trae **24 unidades**.

Usando solo `print()` y operaciones:

1. ¿Cuántas **unidades** llegaron en total?
2. Si se distribuyen entre **3 rutas**, ¿cuántas cajas completas van por ruta? (`//`)
3. ¿Cuántas cajas **sobran**? (`%`)
4. Si las ventas crecen **12% cada mes**, ¿cuántas unidades necesitamos en **6 meses**? (`**`)

**Pista:** Crecimiento compuesto = cantidad × (1 + tasa) `**` meses

```python
# Espacio para resolver
print("1. Unidades totales:", ___)
print("2. Cajas por ruta:",   ___)
print("3. Cajas sobrantes:",  ___)
print("4. En 6 meses:",       ___)
```

```python
# Solución — Recepción de cajas
print(500 * 24)                     # 12 000 unidades
print(500 // 3)                     # 166 cajas por ruta
print(500 % 3)                      # 2 cajas sobrantes
print(500 * 24 * (1 + 0.12) ** 6)   # ≈ 23 700 unidades
```

## 4. Variables: asignar y reutilizar

Una **variable** es un nombre que apunta a un valor en memoria.

- `=` no es "igual a". Significa **asignar**: el valor de la derecha se guarda con el nombre de la izquierda.
- Podemos **reasignar**: `precio = 1.65` actualiza el valor.
- Python distingue mayúsculas: `Precio ≠ precio`.

### Convención: `snake_case`

| Correcto | Evitar |
|----------|--------|
| `precio_unitario` | `p` |
| `costo_por_unidad` | `c1` |
| `margen_pct` | `dato` |
| `total_unidades` | `x` |

```python
precio = 1.50
print(precio)

precio = 1.65
print(precio)
```

### ¿Por qué variables? Comparemos

Si el precio sube a \$1.65, con variables cambiamos **una sola línea** y todo se recalcula.

```python
unidades = 1200
precio = 1.50
costo = 0.85

ingreso = unidades * precio
ganancia = ingreso - unidades * costo

print("Ingreso:", ingreso)
print("Ganancia:", ganancia)
```

### Práctica con variables: Despacho de 3 productos

> Un centro de distribución despacha 3 productos:

| Producto | Precio | Unidades |
|----------|-------:|---------:|
| Coca-Cola 600 ml | \$1.50 | 1 200 |
| Sprite 500 ml | \$1.25 | 800 |
| Dasani 1 L | \$0.90 | 2 000 |

Usando **variables**, calculen e impriman:

1. El **ingreso** de cada producto.
2. El **ingreso total** del despacho (suma de los tres).
3. El **porcentaje** que representa Coca-Cola del total.

**Pista:** Porcentaje = ingreso_coca / ingreso_total × 100

```python
# Espacio para resolver
```

```python
# Solución — Despacho de 3 productos
ingreso_coca = 1.50 * 1200
ingreso_sprite = 1.25 * 800
ingreso_dasani = 0.90 * 2000

ingreso_total = ingreso_coca + ingreso_sprite + ingreso_dasani
pct_coca = ingreso_coca / ingreso_total * 100

print("Coca-Cola:", ingreso_coca)
print("Sprite:", ingreso_sprite)
print("Dasani:", ingreso_dasani)
print("Total:", ingreso_total)
print("% Coca-Cola:", round(pct_coca, 2))
```

### App 1: Cotizador de Ruta *(instructor resuelve)*

> Un vendedor de ruta está con un cliente en Ambato. El cliente quiere **800 unidades de Sprite 500 ml a \$1.25**. La entrega es a **45 km** del centro de distribución y el flete cuesta **\$0.12 por km**. El IVA en Ecuador es **15%**.
>
> Construyamos una herramienta donde cualquier vendedor ingrese los datos de su pedido y obtenga la cotización al instante.

```python
# Solución — Cotizador de Ruta
producto = input("Producto: ")
precio = float(input("Precio unitario: "))
cantidad = int(input("Cantidad: "))
km = float(input("Distancia (km): "))

subtotal = precio * cantidad
flete = km * 0.12
iva = (subtotal + flete) * 0.15
total = subtotal + flete + iva

print("═" * 35)
print("  COTIZACIÓN -", producto)
print("═" * 35)
print("Subtotal:  $", subtotal)
print("Flete:     $", flete)
print("IVA (15%): $", round(iva, 2))
print("─" * 35)
print("TOTAL:     $", round(total, 2))
print("═" * 35)
```

### App 2: ¿Conviene el 3x2? *(estudiantes resuelven)*

> El equipo de marketing quiere lanzar una promoción **"Lleva 3, paga 2"** en Coca-Cola 600 ml. El producto se vende a **\$1.50** y producirlo cuesta **\$0.85**.
>
> El gerente de finanzas pregunta: **¿cuánto margen sacrificamos con esta promoción?**
>
> Creen un programa que pida el precio y el costo de cualquier producto, y compare la ganancia con y sin la promoción 3x2.

**Pista:** Sin promo el cliente paga 3 unidades; con promo paga 2. En ambos casos producimos 3.

```python
# Espacio para resolver
```

```python
# Solución — ¿Conviene el 3x2?
precio = float(input("Precio unitario: "))
costo = float(input("Costo unitario: "))

g_normal = precio * 3 - costo * 3
g_promo = precio * 2 - costo * 3
sacrificio = (g_normal - g_promo) / g_normal * 100

print("Ganancia normal (3 uds):", round(g_normal, 2))
print("Ganancia promo  (3x2): ", round(g_promo, 2))
print("Sacrificio:            ", round(sacrificio, 2), "%")
```

>

### App 4: Proyección para Directorio *(estudiantes resuelven)*

> El directorio de Arca Continental pide una proyección: si la región Quito vende hoy **50 000 unidades/mes** y el crecimiento histórico es **8% mensual**, ¿cuántas unidades venderemos en **12 meses**? ¿Cuánto ingreso representará si el precio promedio es **\$1.35**?
>
> Construyan un programa que reciba ventas actuales, tasa de crecimiento, meses y precio, y entregue la proyección completa.

**Pista:** Crecimiento compuesto: actuales × (1 + tasa/100) `**` meses

```python
# Espacio para resolver
```

```python
# Solución — Proyección para Directorio
uds = int(input("Ventas actuales (uds/mes): "))
tasa = float(input("Crecimiento mensual (%): "))
meses = int(input("Meses a proyectar: "))
precio = float(input("Precio promedio: "))

futuro = uds * (1 + tasa / 100) ** meses
crec = (futuro - uds) / uds * 100

print("═" * 40)
print("  PROYECCIÓN DE VENTAS")
print("═" * 40)
print("Ventas actuales: ", uds, "uds/mes")
print("En", meses, "meses:  ", int(futuro), "uds/mes")
print("─" * 40)
print("Ingreso actual:  $", uds * precio)
print("Ingreso futuro:  $", int(futuro * precio))
print("Crecimiento:     ", int(crec), "%")
print("═" * 40)
```

### Proyecto Integrador: Simulador de Ruta de Distribución

Un camión sale del centro de distribución con carga limitada.  
Debe visitar 6 clientes y decidir en cada parada si puede entregar.

### Reglas por parada:
1. ¿El producto está en el `catalogo`? (`in`)
2. Si no está, reportar `NO DISPONIBLE`
3. Si está pero la `carga` no alcanza, reportar `SIN STOCK`
4. Si se puede, entregar, restar de `carga` y sumar ingreso

### Reportar al final:
- Entregas exitosas
- Pedidos rechazados
- Ingreso total
- Cajas restantes en el camión

````python
# Datos de la ruta
clientes = ["Tienda Sol", "Mini ABC", "Super 10",
            "Bodega LP", "Market 7", "Don Pepe"]
pedidos  = [30, 15, 50, 20, 40, 10]
productos_pedidos = ["Coca-Cola", "Sprite", "Fanta",
                     "Coca-Cola", "Dasani", "Sprite"]

# Carga del camion
catalogo = ["Coca-Cola", "Sprite", "Dasani"]
carga = 80
precio_caja = 12.50

entregas_exitosas = 0
rechazados = 0
ingreso_total = 0.0

for i in range(len(clientes)):
    cliente = clientes[i]
    pedido = pedidos[i]
    producto = productos_pedidos[i]

    if producto not in catalogo:
        print(f"{cliente}: NO DISPONIBLE ({producto})")
        rechazados += 1
    elif pedido > carga:
        print(f"{cliente}: SIN STOCK ({pedido} cajas)")
        rechazados += 1
    else:
        ingreso = pedido * precio_caja
        carga -= pedido
        ingreso_total += ingreso
        entregas_exitosas += 1
        print(f"{cliente}: ENTREGADO | {producto} | {pedido} cajas | ingreso = {ingreso}")

print()
print("Entregas exitosas:", entregas_exitosas)
print("Pedidos rechazados:", rechazados)
print("Ingreso total:", ingreso_total)
print("Cajas restantes:", carga)
````

### Creación y acceso

Un diccionario guarda pares **clave: valor**. Se accede por clave, no por índice.

```
    Clave          Valor
  ─────────    ──────────────
  "nombre"  →  "Tienda Sol"
  "pedido"  →  30
  "producto" → "Coca-Cola"
  "zona"    →  "Norte"
```

````python
cliente = {
    "nombre": "Tienda Sol",
    "pedido": 30,
    "producto": "Coca-Cola",
    "zona": "Norte"
}

print("Nombre:", cliente["nombre"])
print("Pedido:", cliente["pedido"])
````

### Acceso seguro con `.get()`

````python
producto = {
    "nombre": "Coca-Cola 600ml",
    "precio": 1.50,
    "stock": 200
}

# .get() no da error si la clave no existe
print(producto.get("precio"))
print(producto.get("color", "N/A"))  # Default si no existe

# Esto daría KeyError:
# print(producto["color"])
````

### Modificar, agregar y eliminar

````python
producto = {
    "nombre": "Coca-Cola 600ml",
    "precio": 1.50,
    "stock": 200
}

# Modificar un valor existente
producto["precio"] = 1.65

# Agregar una clave nueva
producto["categoria"] = "Bebidas"

# Eliminar una clave
del producto["stock"]

for k, v in producto.items():
    print(f"  {k}: {v}")
````

### Métodos esenciales e iteración

| Método | Devuelve |
|--------|----------|
| `.keys()` | Todas las claves |
| `.values()` | Todos los valores |
| `.items()` | Pares (clave, valor) |
| `.get(k, d)` | Valor de `k`, o `d` si no existe |
| `.pop(k)` | Elimina y devuelve valor de `k` |

````python
producto = {
    "nombre": "Sprite 500ml",
    "precio": 1.25,
    "stock": 150
}

print("Claves:", list(producto.keys()))
print("Valores:", list(producto.values()))
print()

# .items() es la forma más común de iterar
for clave, valor in producto.items():
    print(f"  {clave}: {valor}")
````

## 2. Funciones

### El problema: el código se repite

````python
# Sin funciones — copiar y pegar
ingreso1 = 30 * 12.50
print("Tienda Sol:", ingreso1)

ingreso2 = 50 * 12.50
print("Super 10:", ingreso2)

ingreso3 = 40 * 12.50
print("Market 7:", ingreso3)
````

````python
# Con función — se escribe una vez
def calcular_ingreso(nombre, cajas):
    ingreso = cajas * 12.50
    print(f"{nombre}: ${ingreso:.2f}")
    return ingreso

calcular_ingreso("Tienda Sol", 30)
calcular_ingreso("Super 10", 50)
calcular_ingreso("Market 7", 40)
````

### Anatomía de una función

```
def  calcular_ingreso  (cajas, precio):     ← definición
 │        │                  │
 │     nombre            parámetros
 │
 └─ palabra clave

    ingreso = cajas * precio               ← cuerpo (indentado)
    return ingreso                          ← salida (opcional)
```

- `def` inicia la definición
- Los **parámetros** reciben datos al llamar
- `return` devuelve el resultado. Sin él, la función devuelve `None`

````python
def calcular_ingreso(cajas, precio):
    ingreso = cajas * precio
    return ingreso

resultado = calcular_ingreso(30, 12.50)
print("Ingreso:", resultado)
````

### `return` vs `print` — No son lo mismo

- `print()` muestra texto en pantalla — no se puede reutilizar
- `return` entrega un valor que el programa puede guardar, operar o pasar a otra función

````python
# Solo imprime — no se puede reutilizar
def mostrar_total(cajas, precio):
    print(cajas * precio)

resultado = mostrar_total(30, 12.50)
print("Guardado:", resultado)  # None!
````

````python
# Retorna — el valor se puede usar después
def calcular_total(cajas, precio):
    return cajas * precio

resultado = calcular_total(30, 12.50)
print("Guardado:", resultado)
print("Con IVA:", resultado * 1.12)
````

### Parámetros con valor por defecto

Un parámetro puede traer un valor listo para usar si no se envía otro al llamar la función.

````python
def calcular_ingreso(cajas, precio=12.50):
    return cajas * precio

print(calcular_ingreso(30))
print(calcular_ingreso(30, 15.00))
````

### Alcance de Variables (scope)

Las variables creadas **dentro** de una función son **locales**: solo existen ahí. Si quieren usar ese valor afuera, lo más claro es devolverlo con `return`.

````python
def duplicar(x):
    resultado = x * 2
    return resultado

print(duplicar(5))
# print(resultado)  # NameError: no existe fuera de la funcion
````

### ¿Una función puede cambiar algo fuera de su scope?

Sí, pero depende de lo que recibe y de lo que hace con eso.

Si una función **reasigna** una variable local, no cambia la de afuera. Pero si recibe una **lista** o un **diccionario** y lo modifica, ese cambio sí puede verse afuera porque sigue siendo el mismo objeto.

````python
def cambiar_numero(x):
    x = x + 10

valor = 5
cambiar_numero(valor)
print(valor)


def agregar_alerta(alertas):
    alertas.append("PM2.5 alto")

estado = []
agregar_alerta(estado)
print(estado)
````

### Funciones que Procesan Colecciones

Las funciones pueden recibir listas o diccionarios como parámetros. También pueden retornar varios valores separados por coma.

````python
def resumen_ventas(ventas, meta):
    total = 0
    sobre_meta = 0
    for v in ventas:
        total += v
        if v >= meta:
            sobre_meta += 1
    return total, sobre_meta

datos = [150, 80, 200, 45, 310, 60]
t, s = resumen_ventas(datos, 100)
print(f"Total: {t}")
print(f"Sobre meta: {s}")
````

### Funciones + Diccionarios

Las funciones pueden recibir y procesar diccionarios. Una función puede tomar un diccionario, calcular algo y devolver otro diccionario con resultados.

````python
def procesar_pedido(cliente, precio_caja):
    ingreso = cliente["pedido"] * precio_caja
    return {
        "nombre": cliente["nombre"],
        "ingreso": ingreso,
        "estado": "OK" if cliente["activo"] else "INACTIVO"
    }

cliente = {"nombre": "Tienda Sol", "pedido": 30, "activo": True}
r = procesar_pedido(cliente, 12.50)
print("nombre:", r["nombre"])
print("ingreso:", r["ingreso"])
print("estado:", r["estado"])
````

```python
def calcular_promedio(lista):
    return sum(lista) / len(lista)

def clasificar(valor, umbral):
    if valor >= umbral:
        return "Alto"
    return "Bajo"

def contar_por_clase(lista, umbral):
    altos = 0
    bajos = 0
    for v in lista:
        if clasificar(v, umbral) == "Alto":
            altos += 1
        else:
            bajos += 1
    return altos, bajos

print(calcular_promedio(ventas))      # 140.83...

print(contar_por_clase(ventas, 100))  # (3, 3)
```

<details><summary><b>Ver solución — Pipeline</b></summary>

```python
def filtrar_zona(empleados, zona):
    resultado = []
    for e in empleados:
        if e["zona"] == zona:
            resultado.append(e)
    return resultado

def total_ventas(lista):
    total = 0
    for e in lista:
        total += e["ventas"]
    return total

norte = filtrar_zona(empleados, "Norte")
print(f"Empleados Norte: {len(norte)}")
print(f"Ventas Norte: ${total_ventas(norte):,.2f}")
print(f"Ventas Total: ${total_ventas(empleados):,.2f}")
```

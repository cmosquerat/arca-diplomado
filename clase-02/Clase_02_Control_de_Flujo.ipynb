## 1. Comparaciones y Booleanos

Una **comparación** evalúa una relación entre dos valores y devuelve `True` o `False`.

| Operador | Significado | Ejemplo |
|:--------:|-------------|--------:|
| `==` | igual a | `stock == 150` |
| `!=` | distinto de | `ruta != "Sur"` |
| `>` | mayor que | `margen > 0.20` |
| `<` | menor que | `stock < minimo` |
| `>=` | mayor o igual | `precio >= 1.50` |
| `<=` | menor o igual | `dias <= 30` |

> **Cuidado:** `==` compara, `=` asigna. No confundirlos.

````python
stock   = 150
pedido  = 200
minimo  = 100
precio  = 1.50

print(stock >= pedido)    # ¿Hay stock suficiente para el pedido?
print(stock >= minimo)    # ¿Supera el mínimo de bodega?
print(precio == 1.50)     # ¿El precio es exactamente 1.50?
print(precio != 1.25)     # ¿El precio no es 1.25?
````

### Operadores Lógicos

Combinan condiciones para preguntas más complejas.

| Operador | Resultado |
|----------|-----------|
| `A and B` | `True` solo si **ambos** son `True` |
| `A or B` | `True` si **al menos uno** es `True` |
| `not A` | Invierte: `True` → `False` |

````python
hay_stock         = True
tiene_credito     = False
cliente_bloqueado = False

# and: ambas condiciones deben cumplirse
print(hay_stock and tiene_credito)

# or: basta con que una se cumpla
print(hay_stock or tiene_credito)

# not: invierte el booleano
print(not cliente_bloqueado)
````

## 3. Condicionales: `if / elif / else`

Los booleanos dicen verdadero/falso. Los condicionales dicen **qué hacer** según ese resultado.

```
if condicion_1:
    bloque_1          # se ejecuta si condicion_1 es True
elif condicion_2:
    bloque_2          # se ejecuta si condicion_2 es True (y la 1 fue False)
else:
    bloque_por_defecto
```

- La **indentación** (4 espacios) define qué pertenece a cada bloque.
- Los **dos puntos** `:` al final de cada condición son obligatorios.
- Python evalúa de **arriba hacia abajo** y se detiene en la primera condición verdadera.

````python
# if / else — dos caminos
stock  = 45
minimo = 50

if stock < minimo:
    print("ALERTA: reabastecer")
else:
    print("Stock suficiente")
````

````python
# if / elif / else — múltiples caminos
margen = 0.18

if margen >= 0.25:
    clasificacion = "Rentable"
elif margen >= 0.15:
    clasificacion = "En riesgo"
else:
    clasificacion = "Crítica"

print("Clasificación:", clasificacion)
````

## 5. Listas

Una **lista** guarda múltiples valores bajo un mismo nombre.

- Los elementos van entre **corchetes** `[]` separados por comas.
- Puede contener **cualquier tipo**: números, texto, booleanos.
- El primer elemento tiene índice **0** (no 1).
- Índices negativos cuentan desde el final: `-1` es el último elemento.

````python
rutas = ["Guayaquil Norte", "Guayaquil Sur", "Manta", "Cuenca"]

# len() — cuántos elementos tiene la lista
print(len(rutas))

# Acceder por índice
print(rutas[0])    # Primero
print(rutas[-1])   # Último
````

````python
precios = [1.50, 1.25, 0.90, 1.75]

# Slicing — subconjunto
print(precios[1:3])   # índices 1 y 2 (no incluye el 3)

# Modificar un elemento
precios[0] = 1.65
print(precios)
````

## 7. Ciclos `for`

Un ciclo `for` recorre cada elemento de una lista y ejecuta un bloque de código por cada uno.

```python
for variable_temporal in lista:
    # bloque que se repite
```

- `variable_temporal` toma el valor de cada elemento en cada vuelta.
- El bloque indentado se ejecuta una vez por elemento.
- El nombre de la variable temporal es libre.

````python
# for item in lista
productos = ["Coca-Cola", "Sprite", "Dasani"]

for producto in productos:
    print("Procesando:", producto)
````

````python
# for con range() — contar posiciones
rutas    = ["GYE Norte", "Manta", "Cuenca"]
ingresos = [15200, 12400, 6300]

for i in range(len(rutas)):
    print(rutas[i], "->", ingresos[i])
````

### El Patrón Acumulador

Para sumar o contar sobre una lista:
1. **Antes del ciclo:** inicializar el acumulador en cero.
2. **Dentro del ciclo:** sumar el valor actual.
3. **Después del ciclo:** el acumulador tiene el total.

````python
pedidos = [120, 45, 200, 30, 180, 95, 60, 75]

# Sumar todos los pedidos
total = 0
for cantidad in pedidos:
    total += cantidad          # equivale a: total = total + cantidad

print("Total de cajas:", total)

# Contar pedidos grandes
pedidos_grandes = 0
for cantidad in pedidos:
    if cantidad >= 100:
        pedidos_grandes += 1

print("Pedidos >= 100 cajas:", pedidos_grandes)
````

### `for` + `if` — La combinación poderosa

El `for` recorre; el `if` filtra. Juntos procesan **y** deciden.

````python
pedidos          = [120, 45, 200, 30, 180, 95, 60, 75]
minimo_despacho  = 100

for cantidad in pedidos:
    if cantidad >= minimo_despacho:
        estado = "DESPACHAR"
    else:
        estado = "REVISAR"
    print(cantidad, "->", estado)
````

## 9. Proyecto Aplicado: Clasificador Automático de Pedidos

El área de logística necesita priorizar la preparación de pedidos.
Clasificamos cada uno por **urgencia** y **rentabilidad**.

**Reglas:**

| Criterio | Valor |
|----------|-------|
| Prioridad URGENTE | `urgente = True` |
| Rentabilidad Alta | margen ≥ 30% |
| Rentabilidad Media | 15% ≤ margen < 30% |
| Rentabilidad Baja | margen < 15% |

````python
# Proyecto Aplicado — Clasificador de pedidos
productos  = ["Coca-Cola 600ml", "Sprite 500ml", "Dasani 1L", "Fanta 400ml", "Agua 500ml"]
cantidades = [200, 150, 300, 100, 500]
precios    = [1.50, 1.25, 0.90, 1.10, 0.80]
costos     = [0.85, 0.78, 0.55, 0.72, 0.70]
urgentes   = [True, False, True, False, True]

print("=== CLASIFICADOR DE PEDIDOS ===")
ingreso_total    = 0
pedidos_urgentes = 0

for i in range(len(productos)):
    ingreso = cantidades[i] * precios[i]
    margen  = (precios[i] - costos[i]) / precios[i]
    ingreso_total += ingreso

    # Prioridad
    if urgentes[i]:
        prioridad = "URGENTE"
        pedidos_urgentes += 1
    else:
        prioridad = "NORMAL "

    # Rentabilidad
    if margen >= 0.30:
        rentabilidad = "Alta "
    elif margen >= 0.15:
        rentabilidad = "Media"
    else:
        rentabilidad = "Baja "

    pct = round(margen * 100, 1)
    print(f"{productos[i]}: {prioridad} | {rentabilidad} | {pct}%")

print(f"\nIngreso total:    $ {ingreso_total:.2f}")
print(f"Pedidos urgentes: {pedidos_urgentes}")
````
